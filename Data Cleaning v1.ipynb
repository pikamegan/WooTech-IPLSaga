```python
retained = [0]*129
auct["Retained"] = retained
```

```python
def is_retained(string):
    string = str(string).split(" ")
    if string[-1] == "(retained)":
        return 1
    else:
        return 0
        
auct["Retained"] = auct["Name"].apply(is_retained)
```

```python
def rename(string):
    string = str(string).split(" ")
    new_string = ""
    if len(string) == 2:
        new_string += string[1]
    elif len(string) == 3 and string[2] != "(retained)":
        new_string += string[2]
    elif len(string) == 3 and string[2] == "(retained)":
        new_string += string[1]
    elif len(string) == 4:
        new_string += string[3]
    return new_string
    
auct["Name"] = auct["Name"].apply(rename)
list(auct["Name"])
```

```python
def surname(string):
    string = str(string).split(" ")
    new_string = ""
    if len(string) == 2:
        new_string += string[1]
    return new_string
```

```python
delivs["batsman"] = delivs["batsman"].apply(surname)
list(delivs["batsman"])
```

```python
delivs["non_striker"] = delivs["non_striker"].apply(surname)
list(delivs["non_striker"])
```

```python
delivs["bowler"] = delivs["bowler"].apply(surname)
list(delivs["bowler"])
```
